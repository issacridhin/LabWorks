```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from skimage import exposure
from skimage import feature
```

```python
#image 1
image = cv2.imread("/content/Sonic.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

#image 2
image2 = cv2.imread("/content/pixelblock.jpg")
image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)

#image 3
image3 = cv2.imread("/content/watermelon.jpg")
image3 = cv2.cvtColor(image3, cv2.COLOR_BGR2RGB)
```

```python
#Gray Image reading
gray= cv2.imread('/content/Sonic.jpg',0)
gray2= cv2.imread('/content/pixelblock.jpg',0)
gray3= cv2.imread('/content/watermelon.jpg',0)
```

```python
plt.figure(figsize=(14,12))

plt.subplot(1,3,1)
plt.title("Image 1")
plt.imshow(gray,cmap='gray')
plt.axis("off")

plt.subplot(1,3,2)
plt.title("Image 2")
plt.imshow(gray2,cmap='gray')
plt.axis("off")

plt.subplot(1,3,3)
plt.title("Image 3")
plt.imshow(gray3,cmap='gray')
plt.axis("off")

plt.show()
```

Point extraction

```python
# Create a SIFT object
sift = cv2.xfeatures2d.SIFT_create()

# Detect keypoints and descriptors
keypoints, descriptors = sift.detectAndCompute(gray, None)
keypoints2, descriptors2 = sift.detectAndCompute(gray2, None)
keypoints3, descriptors3 = sift.detectAndCompute(gray3, None)
```

```python
# Adjust circle size and color as desired
for kp in keypoints:
  x, y = kp.pt
  cv2.circle(image, (int(x), int(y)), 5, (0, 255, 0), -1)  # Green circles

for kp in keypoints2:
  x, y = kp.pt
  cv2.circle(image2, (int(x), int(y)), 5, (0, 255, 0), -1)  # Green circles

for kp in keypoints3:
  x, y = kp.pt
  cv2.circle(image3, (int(x), int(y)), 5, (0, 255, 0), -1)  # Green circles
```

```python
plt.figure(figsize=(14,12))

plt.subplot(1,3,1)
plt.title("Points in Image 1")
plt.imshow(image,cmap='gray')
plt.axis("off")

plt.subplot(1,3,2)
plt.title("Points in Image 2")
plt.imshow(image2,cmap='gray')
plt.axis("off")

plt.subplot(1,3,3)
plt.title("Points in Image 3")
plt.imshow(image3,cmap='gray')
plt.axis("off")

plt.show()
```

Line extraction

```python
lsd = cv2.createLineSegmentDetector(0)
```

```python
#Detect lines in the image
lines = lsd.detect(gray)[0]
lines2 = lsd.detect(gray2)[0]
lines3 = lsd.detect(gray3)[0]
```

```python
#Draw detected lines in the image
drawn_img = lsd.drawSegments(gray,lines)
drawn_img2 = lsd.drawSegments(gray2,lines2)
drawn_img3 = lsd.drawSegments(gray3,lines3)
```

```python
plt.figure(figsize=(14,12))

plt.subplot(1,3,1)
plt.title("Lines in Image 1")
plt.imshow(drawn_img,cmap='gray')
plt.axis("off")

plt.subplot(1,3,2)
plt.title("Lines in Image 2")
plt.imshow(drawn_img2,cmap='gray')
plt.axis("off")

plt.subplot(1,3,3)
plt.title("Lines in Image 3")
plt.imshow(drawn_img3,cmap='gray')
plt.axis("off")

plt.show()
```

Edge Extraction

```python
def detect_edges(image):
    edges = cv2.Canny(image, 100, 200)
    return edges
```

```python
edges_image = detect_edges(gray)
edges_image2 = detect_edges(gray2)
edges_image3 = detect_edges(gray3)
```

```python
plt.figure(figsize=(14,12))

plt.subplot(1,3,1)
plt.title("Edges in Image 1")
plt.imshow(edges_image,cmap='gray')
plt.axis("off")

plt.subplot(1,3,2)
plt.title("Edges in Image 2")
plt.imshow(edges_image2,cmap='gray')
plt.axis("off")

plt.subplot(1,3,3)
plt.title("Edges in Image 3")
plt.imshow(edges_image3,cmap='gray')
plt.axis("off")

plt.show()
```
